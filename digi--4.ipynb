```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('census_data.csv')

data.head()
```

```python
dav_data = data[data['county'] == 370] # assuming Davidson County, Nashville

years_of_interest = [1820, 1830, 1840]
dav_years = dav_data[dav_data['year'].isin(years_of_interest)]
population_by_year = dav_years.groupby('year')['ntotal'].sum().reset_index()

plt.figure(figsize=(8,6))
plt.plot(population_by_year['year'], population_by_year['ntotal'], marker='o', linestyle='-')
plt.title("Davidson County Population (1820-1840)")
plt.xlabel("Year")
plt.ylabel("Total Population")
plt.grid(True)
plt.show()
```
